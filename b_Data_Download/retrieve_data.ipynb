This Jupyter Notebook scrapes CitiBike trip data urls and then downloads the files

```python
# Import Dependencies
import pandas as pd 
import requests
import zipfile
import io
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Function to download and extract zip files
def download_zip(zip_file_url,destination_directory):
    # Code snippet from https://stackoverflow.com/questions/9419162/download-returned-zip-file-from-url 
    # Written by yoavram and Kamran Kausar
    r = requests.get(zip_file_url)
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall(destination_directory)
```

```python
# Visit data download page
url = 'https://s3.amazonaws.com/tripdata/index.html'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Collect all zip file URLs
element = browser.links.find_by_partial_href('zip')

# Retrieve download each zip file
for i in range(1,len(element)):
    zip_file_url = element[i]['href']
    print("Downloading: " + zip_file_url) 
    download_zip(zip_file_url,'.')
```

```python
browser.quit()
```
